```python
from captcha.image import ImageCaptcha
import string
import random
from matplotlib import pyplot as plt
from matplotlib import image as mpimg

def generate_captcha():
    # initializing size of string
    N = 7

    # using random.choices()
    # generating random strings
    res = ''.join(random.choices(string.ascii_uppercase + string.digits, k=N))

    #print(type(res))
    #print("The Captcha Image must contain: {}".format(res))

    # Create an image instance of the given size
    image = ImageCaptcha(width = 280, height = 90)

    # Image captcha text
    captcha_text = res

    # generate the image of the given text
    data = image.generate(captcha_text)

    # write the image on the given file and save it
    image.write(captcha_text, 'CAPTCHA.png')
    
    plt.title("Captcha Image")
    #plt.xlabel("X pixel scaling")
    #plt.ylabel("Y pixels scaling")
    plt.grid(False)
    plt.axis('off')

    image = mpimg.imread("CAPTCHA.png")
    plt.imshow(image)
    plt.show()

    get_user_auth = input("Enter the Captcha: ").upper()
    print("User input: {}".format(get_user_auth))
    if get_user_auth != res:
        print("\nAuthentication Failed!\nGenerated Captch: {}\nUser Answer: {}".format(res, get_user_auth))
    else:
        print("\nIt matched!\nYou are not a Robot.")

generate_captcha()
```
